Having

 $$\nabla E = \rho / \varepsilon_0 \text{ , } \nabla H = 0 \text{ , } \nabla \times E = - \partial_t B \text{ , } \nabla \times H = J + \partial_t D \text{ , } D = \varepsilon_0 E + P \text{ , } B = \mu_0 H + M$$

 we find


$$ \nabla \times \nabla \times E = \Delta E - \nabla (\nabla E) = - \partial_t (\nabla \times B) $$


and having linear materials, as well as no sources present, we find


$$ \Delta E(\vec{r}, t) + \mu(\vec{r}) \epsilon(\vec{r}) \cdot \partial_t^2 E(\vec{r}, t) = 0 $$


As we assume that $\vec{E}(x,y,t) = Re[E_z(x,y) \cdot \vec{e}_z \cdot e^{i\omega t}]$ (ToDo : How can we assume that -  or do we switch to frequency domain?), we find


$$ (\partial_x^2 + \partial_y^2) E_z + \mu\varepsilon(x,y) \cdot \omega^2 E_z = 0$$


If we additionally assume that the EM-field is propagating in the $x$-direction, we find $\vec{E}(x,y,t) = Re(E_{z0}(y) \cdot e^{-i\gamma x} \cdot \vec{e}_z \cdot e^{-i \omega t})$ and therefore $ \partial_x E_z = - i \gamma E_{z0} \cdot e^{-i \gamma x}$ as well as $\partial_x^2 E_z = - \gamma^2 E_{z0} \cdot e^{-i \gamma x} $ (ToDo : Again how do these assumptions come by?). Plugging in, we find


$$ - \gamma^2 E_{z0} \cdot e^{-i \gamma x} + \partial_y^2 E_{z0} \cdot e^{-i \gamma x} + \mu\varepsilon(x,y) E_{z0} \cdot e^{-i \gamma x} = 0 \rightarrow \partial_y^2 E_{z0} + (\mu\varepsilon(x,y) - \gamma^2) E_{z0} = 0 \rightarrow \partial_y^2 E_{z0} + k^2 E_{z0} = 0$$


where we defined $k^2 = \mu\varepsilon \omega^2 - \gamma^2$ (ToDo : We need $k^2 > 0$ to get the right solution - can this be violated?). Solving the second order ODE yields


$$ E_{z0}(y) = A \cdot cos(k \cdot y) + B \cdot sin(k \cdot y) $$


and combining the results we find


$$ E_z(x,y) = A \cdot cos(k \cdot y) \cdot e^{-i \gamma x} + B \cdot sin(k \cdot y) \cdot e^{-i \gamma x} $$


We will therefore find even and odd modes. As the waveguide is terminated by perfect electric conductors, we have that $E_z = 0 \ @ \ y = \pm d/2$ and find for the even modes

$$ cos(k \pm d/2) = 0 \rightarrow k \pm d/2 = \pm \pi/2 \pm 2n\pi \rightarrow k = \frac{(2n + 1)\pi}{d} \text{ for } n = 0, 1, \dots$$

and for odd modes

$$ sin(k \pm d/2) = 0 \rightarrow k \pm d/2 = n \pi \rightarrow k = \frac{2n \pi}{d} \text{ for } n = 1, 2, \dots$$

If we have that $\gamma \in \mathbb{C}$, we find evanescent modes. We want $\gamma \in \mathbb{R}$, we find

$$ \gamma = \sqrt{\mu\varepsilon \omega^2 - k^2} \rightarrow w \geq \frac{k}{\sqrt{\mu\varepsilon}} = \frac{kc}{\eta} = k v $$

where we call the smallest $\omega$ such that the above condition is fullfilled the cut off frequency. We are given the following values

$$ d = 0.02 \ [m] $$

$$ \varepsilon = 8.85 \cdot 10^{-12} \ [F/m] $$

$$ \mu = 12.56 \cdot 10^{-7} \ [Tm/A] $$

Thus, for the fundamental even mode, we find

$$ k = \frac{\pi}{d} = 157 \rightarrow f \geq 7.5 \cdot 10^9 \ [Hz] \leftrightarrow \text{ First dominant even mode - Single mode propagation range - ToDo }$$

We are also given general boundary conditions where the normal vector $\vec{n}$ is neither in the $x$ or the $y$ direction.

Having $f : \mathbb{R}^2 \to \mathbb{R}, \vec{x} \mapsto f(\vec{x})$ with $- \nabla (\alpha \nabla f) - \beta f = g \text{ for } \vec{x} \in \Omega$ subject to $f(\vec{x}) = p \text{ for } \vec{x} \in \partial_D \Omega$ and $\partial_{\vec{n}}f(\vec{x}) = q \text{ for } \vec{x} \in \partial_N \Omega$ where $\partial \Omega = \partial_D \Omega \cup \partial_N \Omega$, with the weightet residual approach (multiply equation by $w$), through integration by parts ($ \partial_x(fg) = \partial_x f g + f \partial_x g$) and using the divergence theorem 

$$ \int\limits_{\Omega} \nabla_{\vec{x}} (\nabla_{\vec{x}} f) d\Omega = \int\limits_{\partial \Omega} \nabla_{\vec{x}} f d\vec{A} = \int\limits_{\partial \Omega} \nabla_{\vec{n}} f dA = \int\limits_{\partial_N \Omega} q dA $$

where it was used $\nabla_{\vec{x}} g \cdot \vec{n} = \nabla_{\vec{n}} g$, we find

$$ \int\limits_{\Omega} \alpha \nabla w \nabla f d\Omega + \int\limits_{\Omega} \beta w f d\Omega - \int\limits_{\Omega} w g d\Omega - \int\limits_{\partial \Omega} w \nabla_{\vec{n}} f dA = 0 $$

by switching to the global representation, we find

$$ \sum\limits_{e = 1}^{N_e} \int\limits_{\Omega^{(e)}} \alpha^{(e)} \nabla w \nabla f^{(e)} d\Omega^{(e)} + \sum\limits_{e = 1}^{N_e} \int\limits_{\Omega^{(e)}} \beta^{(e)} w f^{(e)} d\Omega^{(e)} - \sum\limits_{e=1}^{N_e} \int\limits_{\Omega} w g^{(e)} d\Omega - \sum\limits_{\substack{e \text{ such that }\\ \partial\Omega^{(e)} \in \partial_N\Omega}} \int\limits_{\partial \Omega^{(e)}} w q^{(e)} dA^{(e)} = 0 $$

switching to the global notation by using

$$ \sum\limits_{e=1}^{N_e} f^{(e)} = \sum\limits_{n = 1}^{N_n} f_n N_n $$

reversing the sums and letting $w = N_m$, we find

$$ \sum\limits_{n = 1}^{N_n} f_n \left( \sum\limits_{\substack{e \in \\ Supp(N_n) \\ Supp(N_m)}} \int\limits_{\Omega^{(e)}} \alpha^{(e)} \nabla N_m \nabla N_n d\Omega^{(e)} \right) + \sum\limits_{n=1}^{N_n} f_n \left( \sum\limits_{\substack{e \in \\ Supp(N_n) \\ Supp(N_m)}} \int\limits_{\Omega^{(e)}} \beta^{(e)} N_m N_n d\Omega^{(e)} \right) = \sum\limits_{\substack{e \in \\ Supp(N_m)}} \left[ \int\limits_{\Omega^{(e)}} N_m g^{(e)} d\Omega^{(e)} \right] + \sum\limits_{\substack{e = 1 \\ \partial\Omega^{(e)} \in \partial_N\Omega \\ Supp(N_m)}}  \left[ \int\limits_{\partial \Omega^{(e)}} N_m q^{(e)} dA^{(e)} \right] \ \forall \ m \in \{1, 2, \dots N_n\}$$

evaluating the terms in the brackets for the different nodes leads to

$$ A \vec{f} = \vec{b} $$

subsequently enforcing the boundary conditions results in a solvable LSE yielding the resulting approximated unknown function.
